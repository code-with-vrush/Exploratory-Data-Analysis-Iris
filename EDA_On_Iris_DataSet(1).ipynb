#<span style='font-size:80px;'>&#9734;</span><u>what is EDA</u>?

<p><span style='font-size:50px;'>&#8618;</span><b>Exploratory data analysis (EDA)</b> is used by data scientists to analyze and investigate data sets and summarize their main characteristics, often employing data visualization methods.

EDA helps determine how best to manipulate data sources to get the answers you need, making it easier for data scientists to discover patterns, spot anomalies, test a hypothesis, or check assumptions.

EDA is primarily used to see what data can reveal beyond the formal modeling or hypothesis testing task and provides a provides a better understanding of data set variables and the relationships between them. It can also help determine if the statistical techniques you are considering for data analysis are appropriate. Originally developed by American mathematician John Tukey in the 1970s, EDA techniques continue to be a widely used method in the data discovery process today.</P>

Whereas, a DataFrame is flexible in terms of size and number of columns and rows as well as the type of data being held in it and the data can be accessed using referencing through row and column names while, Series is a one-dimensional single array with varied types of data and the referencing of the data is done through its label.

Pandas contains many data manipulation operations including merge/join the given datasets, reshape the data and group by operation for an aggregation. It also contains the sound data preprocessing features including handling of the missing data, filtering and type conversion. As well, the computation with time series is another one of the strong sides of the Pandas, because the library has specific functions that are useful for resampling the data, shift the time, use the rolling windows. They include operations that allow one to conveniently read from and write to various formats such as CSV, Excel, SQL, and JSON.

Pandas seamlessly integrates well with the packages that exist on Python with regard to data analysis these include; NumPy which is efficient in numerical calculation, Matplotlib which is efficient in data visualization and SciPy which is efficient in general computation in science and engineering. Due to this, it remains among the most crucial functions as data pre-processing, cleaning, visualization, and even modeling.

````python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Adjust space between plots
plt.subplots_adjust(hspace=0.4, wspace=0.4)

# Violin plot for Sepal Length
sns.violinplot(ax=axes[0, 0], x='species', y='sepal_length', data=iris_df, palette="crest")
axes[0, 0].set_title('Violin plot of Sepal Length by Species')

# Violin plot for Sepal Width
sns.violinplot(ax=axes[0, 1], x='species', y='sepal_width', data=iris_df, palette="crest")
axes[0, 1].set_title('Violin plot of Sepal Width by Species')

# Violin plot for Petal Length
sns.violinplot(ax=axes[1, 0], x='species', y='petal_length', data=iris_df, palette="crest")
axes[1, 0].set_title('Violin plot of Petal Length by Species')

# Violin plot for Petal Width
sns.violinplot(ax=axes[1, 1], x='species', y='petal_width', data=iris_df, palette="crest")
axes[1, 1].set_title('Violin plot of Petal Width by Species')

# Display the plots
plt.show()
````

````python
p = (
    so.Plot(iris_df, x="petal_length", y="petal_width", color="species")
    .facet(col="species")
    .add(so.Dots(), so.Jitter(0.10))
    .label(
        x="Petal Length",
        y="Petal Width"
    )
    .layout(size=(15,6))
)
p.show()
````

````python
# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Set colors and markers
colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
markers = {'setosa': 'o', 'versicolor': '^', 'virginica': 's'}

# Plot each species with different colors and markers
for species in iris_df['species'].unique():
    subset = iris_df[iris_df['species'] == species]
    ax.scatter(subset['sepal_length'], subset['sepal_width'], subset['petal_length'],
               color=colors[species], marker=markers[species], label=species, s=100)

# Add labels and legend
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
ax.legend()

plt.title('3D Scatter Plot of Iris Dataset')
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(8, 5))
# Animation function
def animate(i):
    ax.clear()  # Clear the axis for each frame
    sns.scatterplot(data=iris_df, x='sepal_length', y='petal_length', hue='species', ax=ax, palette='crest', legend=False)
    sns.kdeplot(data=iris_df[:i], x='sepal_length', y='petal_length', ax=ax, fill=True, alpha=0.3)
    ax.set_title(f'Frame {i+1}/{len(iris_df)}')

# Create the animation
ani = animation.FuncAnimation(fig, animate, frames=len(iris_df), interval=100, repeat=False)
# Save the animation
ani.save('simple_animated_plot.mp4', writer='ffmpeg', fps=15)
# Show the plot
plt.show()
````
